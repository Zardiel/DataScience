```python
import pandas as pd
```

```python
order = pd.read_csv('orders.csv', sep=';')
order = order.rename(columns={'ID товара': 'Product_ID'}, inplace=False)
order.head(1)
```

```python
product = pd.read_csv('products.csv', sep=';')
product.head(1)
```

```python
# Объедините заданные таблицы в таблицу orders_products, чтобы в результирующей таблице оказалась информация обо всех заказах, 
# но не оказалось информации о продуктах, на которых заказов ещё не поступало. 
orders_products = pd.merge(order, product, on='Product_ID', how='left')
print(orders_products)
```

```python
# Какой покупатель принёс наибольшую суммарную прибыль интернет-магазину за указанный период?
mask1 = orders_products.groupby('ID Покупателя')['Price'].sum()
display(mask1)
```
